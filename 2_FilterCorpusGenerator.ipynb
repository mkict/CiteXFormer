Call bulk search API using register API key using the following parameter:
1. fieldsOfStudy: filter only computer science field paper
2. fields: response field
3. year: paper publish year
4. publicationTypes: only paper accept

then write response to filtered_corpusid_paperid file

```python
import requests
import json
import time
import pandas as pd

#S2O_API_KEY = "Please request your Semantic Scholar API Key follows this link (https://www.semanticscholar.org/product/api#api-key-form)"

base_url = "https://api.semanticscholar.org/graph/v1/paper/search/bulk"
headers = {
    "x-api-key": S2O_API_KEY
}
params = {
    "fieldsOfStudy": "Computer Science",
    "fields": "paperId,corpusId,title,abstract,referenceCount,citationCount,influentialCitationCount,year,publicationDate,authors",
    "year": "2015-",
    "publicationTypes": "Conference"
}

filepath = "/content/drive/Shareddrives/NLP Fall 2023/Dataset/filtered_corpusid_paperid.csv"

response = requests.get(base_url, headers=headers, params=params)
data = json.loads(response.text)
params["token"] = data["token"]

df = pd.DataFrame(data["data"])
df.to_csv(filepath, index=False)

# limit 100 call per minute
if params["token"] is not None:
  for i in range(1000000):
      response = requests.get(base_url, headers=headers, params=params)

      data = json.loads(response.text)
      df = pd.DataFrame(data["data"])
      df.to_csv(filepath, mode='a', header=False, index=False)

      params["token"] = data["token"]
      if params["token"] is None:
        break

      time.sleep(0.3) # 50 call per minute
```
