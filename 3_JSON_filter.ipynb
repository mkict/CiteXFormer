Filter dataset file using corpus id and write to new dataset file

```python
import pandas as pd
import json

base_path = "/content/drive/Shareddrives/NLP Fall 2023/Dataset/"

df = pd.read_csv(base_path + "filtered_corpusid.csv")
corpus_ids = set(df["corpusId"].tolist())

for i in range(1,31):
  filter_data = []
  with open(base_path + "unzip/dataset_"+ str(i) + ".jsonl", "r", encoding="utf-8") as file:
      for j, line in enumerate(file):
          line_json = json.loads(line)
          if line_json["corpusid"] in corpus_ids:
              filter_data.append(line_json)

  with open(base_path + "filter/dataset_"+ str(i) +".json", "w", encoding="utf-8") as f:
      json.dump(filter_data,f)
```
