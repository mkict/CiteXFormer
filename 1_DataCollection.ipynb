```python
import requests
import urllib
import gzip
import shutil
```

We downloaded the full-corpus dataset from the **Semantic Scholar Academic Graph (S2AG)** using the Datasets API with release version 2023-10-19. The API response consists of pre-signed download URLs.

```python
#S2O_API_KEY = "Please request your Semantic Scholar API Key follows this link (https://www.semanticscholar.org/product/api#api-key-form)"

headers = {
    'x-api-key': S2O_API_KEY
}

# Get info about the latest release (updated 2023-10-19)
latest_release = requests.get('http://api.semanticscholar.org/datasets/v1/release/latest/dataset/s2orc/',headers=headers).json()
```

This dataset is partitioned and stored on S3, allowing us to retrieve them by requesting these URLs and fetching all the partitions.

```python
latest_release['files']
```

```python
# Get zip file from url to folder zip
for idx, url in enumerate(latest_release['files']):
  filename = "/content/drive/Shareddrives/NLPDataset/zip/dataset_"+str(idx+1)+".gz"
  print(url)
  urllib.request.urlretrieve(url, filename)

# Extract zip file to folder unzip
for idx in range(len(latest_release['files'])):
  zip_filename = "/content/drive/Shareddrives/NLPDataset/zip/dataset_"+str(idx+1)+".gz"
  unzip_filename = "/content/drive/Shareddrives/NLPDataset/unzip/dataset_"+str(idx+1)+".jsonl"
  with gzip.open(zip_filename, 'rb') as f_in:
    with open(unzip_filename, 'wb') as f_out:
      shutil.copyfileobj(f_in, f_out)
```
