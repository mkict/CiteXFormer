Read filtered_corpusid_paperid and save to corpusList

```python
import pandas as pd

base_path = "/content/drive/Shareddrives/NLP Fall 2023/Dataset/"

corpusList = pd.read_csv(base_path + 'filtered_corpusid_paperid.csv')
corpusList
```

Add the following column:
1. referenceCount
2. citationCount
3. year
4. publicationDate
to main dataset file

```python
corpusId_set = set(corpusList['corpusId'])

for i in range(1,31):
  dataset = pd.read_json(base_path + 'filter/dataset_'+str(i)+'.json')
  for idx,row in dataset.iterrows():
    if row['corpusid'] in corpusId_set:
      dataset.loc[dataset['corpusid'] == row['corpusid'],'referenceCount'] = corpusList.loc[corpusList['corpusId'] == row['corpusid']]['referenceCount'].values[0]
      dataset.loc[dataset['corpusid'] == row['corpusid'],'citationCount'] = corpusList.loc[corpusList['corpusId'] == row['corpusid']]['citationCount'].values[0]
      dataset.loc[dataset['corpusid'] == row['corpusid'],'year'] = corpusList.loc[corpusList['corpusId'] == row['corpusid']]['year'].values[0]
      dataset.loc[dataset['corpusid'] == row['corpusid'],'publicationDate'] = corpusList.loc[corpusList['corpusId'] == row['corpusid']]['publicationDate'].values[0]
  dataset.to_json(base_path + 'data/dataset_'+str(i)+'.json')
```
